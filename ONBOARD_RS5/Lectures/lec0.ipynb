### **영상이 없는 콘텐츠**

---

Fundamental의 각 노드는 영상이 아닌 텍스트와 코드로 이루어져 있습니다. 보통의 강의와는 달라 놀라셨나요? 😅<br>

아이펠의 콘텐츠는 영상이 있는 것도 있지만 텍스트와 코드로 구성된 것들도 있습니다. Fundamental이 텍스트와 코드로 구성된 콘텐츠 중 하나에요.<br>

어떤 분들은 '전자책도 아니고 이게 무슨 콘텐츠냐?'라고 생각하실 수도 있어요. 영상이 없어도 여러분들이 학습하실 때 부족한 내용이 있거나 이해하기 어렵지 않도록 내용을 충실히 보강하고 있어요. 콘텐츠는 계속 진화 중입니다. 😁

### **콘텐츠 구성**

---

Fundamental은 크게 아래와 같이 구성되어 있습니다.

- 파이썬 기초
- 데이터 분석
- 머신러닝
- 딥러닝 기초 이론
- 딥러닝 심화 이론

여러분들은 각 카테고리별로 여러 개의 노드를 학습하실 거에요. 짧은 기간동안 학습하기 때문에 각 카테고리의 모든 내용을 다루지는 못하고, 일부만 학습하게 될 거에요.<br>

앞으로 더 많은 내용, 더 깊은 내용을 다루기 위해 Fundamental을 이처럼 길지 않게 구성하였어요. 우리의 여정은 Fundamental에서 끝나는 것은 아니니까요.<br>

기초가 부족하다고 생각하신다면 개인적으로 추가 학습을 하시거나 스터디를 통해 다른 분들과 함께 공부해 보는 것을 추천해 드려요. 🙏

### **Fundamental의 학습 가이드**

---

Fundamental은 특별한 과제나 프로젝트는 없습니다. 다만, 여러분의 학습을 돕기 위해 각 스텝마다 다양한 퀴즈와 배운 내용을 총정리하는 종합 퀴즈가 있어요.<br>

처음에는 노드에 과제나 프로젝트가 없고, 각 노드도 텍스트와 코드로 구성되어 있어서 어떻게 학습해야할 지 고민하실 분들이 있을 거에요. 뿐만 아니라 블로그나 공식 문서 링크를 주고 읽어보라고 할 때면 더 당황스럽겠죠?<br>

이제부터 어떻게 공부하면 좋을지 함께 고민해 봐요. 아래의 학습 가이드는 정답이 아니라 이렇게 공부하면 좋을 것 같다는 조언입니다. 공부 방법을 잘 모르겠다는 분들은 아래의 내용을 참고해 주세요!

#### **텍스트**

모든 내용을 한 번에 습득해야 한다는 생각을 버리고 전체 내용을 쭉 읽고 핵심 내용을 정리해 보세요. 기억이 나지 않는 세부 내용은 해당 내용으로 돌아가 다시 읽어 보세요. 중간중간에 있는 퀴즈와 종합 퀴즈를 풀면서 내용을 정리하는 것도 좋아요.

#### **블로그 또는 공식 문서 링크**

텍스트 중간에 있는 블로그 링크는 모두 읽어보라고 삽입한 것은 아니에요. 여러분의 학습 수준에 따라 다양한 정보를 드리고자 추가 학습 자료를 제공한 것이니 필요한 경우에만 링크에 들어가 보세요.

- **꼭 알고 있어야 하는 블로그/공식 문서 :** 꼭 알고 있어야 하는 부분은 퀴즈를 통해 알려드리니, 블로그/공식 문서 내용은 쓱 훑고(skimming) 퀴즈를 통해 알아야 할 부분만 확실히 숙지하고 넘어가세요.

- **참고만 해도 되는 블로그/공식 문서 :** 꼭 알지 않아도 되지만 알고 있으면 좋을 내용을 블로그/공식 문서 링크로 삽입하는 경우가 많아요. 이런 경우에는 퀴즈가 따로 제공되지 않아요. 노드를 모두 학습하고 시간이 남으면 링크를 클릭하여 읽어 보세요.

#### **퀴즈**

위에서 언급했듯 퀴즈에는 각 스텝별마다의 퀴즈와 종합 퀴즈가 있어요. 퀴즈는 보통 예시 답안이 제공되기 때문에 시간이 부족하거나 귀찮으면 답안을 작성하지 않고 그냥 넘어가기 십상인데요, 여러분의 학습을 위해 열심히 만들었으니 퀴즈를 꼭 풀고 넘어가 주세요. 퀴즈를 푸는 것, 역시 학습의 부분이랍니다. 퀴즈를 열심히 풀다 보면 여러분의 지식과 문제 해결 능력이 쑥쑥 성장해 있을 거에요. 🌱

#### **코드**

콘텐츠에는 필요할 때마다 실습을 할 수 있는 코드가 있어요. 코드가 이미 작성되어 있는 곳도 있지만 코드 퀴즈도 있습니다. 코드가 이미 작성된 곳은 [코드 실행]을 누르고 그냥 넘어가지 말고, 코드를 꼼꼼히 살펴 보세요. 코드 셸에서는 코드를 자유롭게 변경할 수 있으니 코드를 다양하게 수정해 보면서 결과를 비교해 보는 것도 좋아요.

코드 퀴즈는 특별한 경우가 아니면 정답을 제공하지 않아요. 하지만 코드 셸 윗 부분의 콘텐츠를 충분히 학습하면 해결할 수 있는 퀴즈를 넣었으니 모른다고 바로 질문하지 말고, 충분한 시간을 갖고 퀴즈를 해결해 보세요. 직접 코드를 작성하는 능력은 개발자에게 필수랍니다!
