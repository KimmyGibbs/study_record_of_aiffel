## 5.1 일반화: 머신러닝의 목표

머신러닝에서 훈련이 반복될수록 훈련데이터의 성능이 향상되고 이는 모델의 `과대적합(overfitting)`을 야기한다.</br>

&nbsp;</br>
**최적화**(optimization)
- 머신러닝 학습 시 가능한 훈련데이터에서 최고의 성능을 얻기위해 모델을 조정하는 것

**일반화**(generalization)
- 훈련된 모델이 이전에 본 적없는 데이터에서 얼마나 잘 수행되는지를 의미
    - 좋은 일반화 성능을 얻는게 궁극적인 머신러닝의 목표이긴 하지만 일반화 성능을 제어할 방법이 없다.

**홀드아웃**(Holdout)
- 데이터셋을 train_data, test_data로 나누는 과정을 의미
> [참고링크](https://m.blog.naver.com/sjc02183/221739648990)

**과소적합**(underfitting)
- 훈련 데이터의 손실이 낮아질수록 테스트 데이터의 손실도 낮아지는 상황
    - 모델의 성능이 계속 발전될 여지가 있음

**과대적합**(overfitting)
- 훈련을 특정 횟수만큼 반복 후, 일반화 성능이 높아지지 않으면서 검증 세트의 성능이 감소되기 시작하는 상황
    - 데이터에 잡음(noise)이 있거나, 불확실성이 존재하거나, 드문 특서이 포함되어 있을 때 특히 과대적합이 발생할 가능성이 높다.

### **불확실한 특성**

문제에 불확실성과 모호성이 있다면 레이블이 완벽하고 깔끔하게 부여된 데이터라 하더라도 잡음이 있을 수 있다.</br>

&nbsp;</br>
분류 작업에서 입력 특성 공간의 일부 영역이 여러 클래스에 동시에 연관된 경우</br
1) 바나나 이미지를 받아서 바나나의 숙성도를 예측하는 모델 (또는 바나나가 썩었는지 아닌지를 여측하는 모델)
    - 이런 category에는 객관적인 경계가 없어 동일한 데이터에 대해서 다른 레이블로 분류될 수 있다.
2) 기압 데이터를 사용해서 내일 비가 올지 아닐지를 여측하는 모델</br>
    - 정확히 동일한 측정값을 얻은 후에도 일정 확률로 이따금 비가 오거나 맑은 하늘이 된다.

특성 공간의 모호한 영역에 너무 확신을 가지면 확률적인 데이터의 경우에서 과대적합이 발생할 수 있다.</br>
> 최적적합을 하려면 개별 데이터 포인트를 무시하고 더 큰 그림을 바라보아야 한다.
