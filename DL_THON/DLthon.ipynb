## **1-2. DLthon 소개**

#### **DLthon이란?**

---

DLthon을 한다고 하는데 DLthon이 뭔가요? 🤷

DLthon이란 딥러닝(DeepLearning)과 마라톤(marathon)의 합성어로 해커톤(해킹+마라톤)에서 착안한 단어입니다.<br>
DLthon은 제한 시간 내에 주어진 데이터를 분석하여 팀별로 주어진 task를 end-to-end로 해결해 보는 시간입니다.<br>
즉, 데이터를 활용한 딥러닝 프로젝트의 파이프라인을 직접 설계하고, 모델의 성과를 확인하는 과정이라고 생각할 수 있습니다.
<br><br>

#### **DLthon의 목적**

---

1. 팀 프로젝트를 경험하기

   - 실무에 들어가기 전 경험하기 가장 어려운 것 중 하나는 바로 협업입니다.
    공동으로 작업을 작성하는 피상적인 협업을 넘어, 각 팀원 여러분이 직접 기획부터 코딩까지 프로젝트 과정을 몸으로 경험하고,
    함께 의견을 모으면서 동등하게 소통해 의사를 결정하는 경험이 되길 바랍니다.

   - 소통과 협업이 강조되는 팀 프로젝트를 경험해 보면, 텍스트로 전달되던 노드만 공부할 때보다 훨씬 입체적이고 실전적인 것들을 경험할 수 있습니다. 😉

2. 딥러닝의 전체 flow를 이해하기

   - 우리가 노드를 통해 딥러닝에 관련된 내용을 학습하고 있지만, 실제 상황과 유사하게 독립된 프로젝트 흐름에 딥러닝을 적용해보는 경험은 어떤 공부보다 중요하고 효율적입니다.

   - DLthon은 초기 기획부터 테스트의 결과에 대한 분석까지 진행되며, 딥러닝에 대해 구석구석 씹.뜯.맛.즐 해볼 수 있습니다.

3. 데이터를 자유롭게 가지고 놀아보기

   - 우리는 앞으로도 계속 여러 종류의 다채로운 데이터셋들을 마주치게 될 겁니다.
    백지에서부터 데이터들을 마음껏 가지고 놀아보면서, 새로운 데이터에 대한 두려움은 잠시 접어두고 나만의 작업 방식을 다듬어가면 어떨까요?

4. 성능 개선을 위해 논리적으로 접근하기

    - 여러분들은 이번 DLthon을 통해 모델의 성능을 올리기 위한 여러가지 전처리 기법, 문제 상황에 알맞은 다양한 모델, 데이터에 최적화된 하이퍼 파라미터 조합 등을 다각적으로 접근하고 논리적으로 탐색할 수 있습니다.

    - 동일한 모델이라고 하더라도 하이퍼 파라미터 조합을 통해 성능이 변할 수 있으며, 이러한 조건들이 우리의 상황에 어떤 영향을 미치는지, 각 특성이 데이터적으로 무엇을 의미하는지 세세하게 고민하고 이해할 수 있습니다.

DLthon을 진행하기 전 위에 나열된 목적들을 되새겨보시고,
앞서 과정들을 통해 배운 내용을 바탕으로, 목표를 설정하고, 협업하며, 회고하는 시간을 가지면서
어제보다 한 걸음 발전한 ‘우리’를 응원하는 시간을 보내시면 좋겠습니다! 화이팅!~🏋️‍♀️

#### **팀 프로젝트 가이드**

---

팀 프로젝트를 경험해본 적 없어서 걱정이 되시나요? 도대체 어떤 것부터 시작하면 좋을까요?
이런 분들을 위해서 팀 프로젝트 가이드를 간단히 소개해 드릴게요 😉

1. **팀 내 역할 분담을 진행합니다.**<br><br>

   - 팀을 이끄는 팀장을 먼저 정하고, 팀원들과 적절하게 역할 분담을 진행합니다.

   - 모든 팀원이 충분히 의견을 교환하는 것이 중요합니다. (특정한 한 분이 하드캐리 하는 것은 DLthon의 취지와 어긋납니다.)

2. **DLthon 기간 동안 일정 계획을 세웁니다.**<br><br>

   - 우리에게 남은 시간 중 구체적으로 어떤 일정을 언제 수행할 건지 팀원들과 함께 계획을 세웁니다.<br> 프로젝트 기간은 제한이 있기 때문에 구체적으로 일정을 세워 진행하는 것이 결과물을 마무리하기 좋습니다.

3. **팀명, 팀 각오, 목표 등을 세웁니다.**<br><br>

   - 팀으로 진행되는 프로젝트이기 때문에 팀명, 팀 각오를 세우면서 우리의 의지를 다져보는건 어떨까요?🔥<br>활동을 시작하기 전에 중간 점검 목표를 설정하면 일정을 잘 맞추고 있는지 확인하기에 편리하고,<br>
   팀원 간에 결집력을 다지고 동기를 유지하는데 유리합니다.

4. **프로젝트를 포트폴리오로써 관리합니다.**<br><br>

   - 우리가 프로젝트 기간 동안 땀흘려 열심히 진행한 과정과 결과를 github, notion 등에 차곡차곡 기록하고 관리한다면,<br> 이후에 개인적인 포트폴리오를 구성하는데 많은 도움이 될 것 같습니다.
